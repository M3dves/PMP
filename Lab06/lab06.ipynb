```python
import numpy as np
from scipy.stats import gamma


# date observate
n_ore = 10
total_apeluri = 180

# PRIORS
# folosim prior flat (fara informatii anterioare)
alpha0 = 0
beta0 = 0

# parametri posteriori

# Conjugarea Gamma-Poisson: posterior tot Gamma
alpha_post = alpha0 + total_apeluri #shape
beta_post = beta0 + n_ore      #rate

# scale = 1/rate
scale_post = 1 / beta_post


# Statistici posteriori
posterior_mean = alpha_post / beta_post                     # media
posterior_variance = alpha_post / (beta_post ** 2)          # varianta
posterior_mode = (alpha_post - 1) / beta_post if alpha_post > 1 else 0  # modul

# Calcul HDI 94%
prob = 0.94

# generam intervale candidate pentru HDI
qs = np.linspace(0, 1 - prob, 100000)
lower_vals = gamma.ppf(qs, a=alpha_post, scale=scale_post)
upper_vals = gamma.ppf(qs + prob, a=alpha_post, scale=scale_post)
widths = upper_vals - lower_vals
min_idx = np.argmin(widths)
hdi_lower = lower_vals[min_idx]
hdi_upper = upper_vals[min_idx]


print(f"Posterior: Gamma(alpha={alpha_post}, beta={beta_post})") #a)
print(f"Media posterioara = {posterior_mean:.3f}")
print(f"Varianta posterioara = {posterior_variance:.3f}")
print(f"Modul posterior (valoarea cea mai probabila pentru lambda)= {posterior_mode:.3f}") #c)
print(f"Interval HDI 94% = [{hdi_lower:.3f}, {hdi_upper:.3f}]") #b)
```
