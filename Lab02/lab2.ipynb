Exercitiul 1:

```python
import pymc as pm
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def experiment():
 urn={"red":3,"blue":4,"black":2}
 face=np.random.randint(0,7)
 if face in (2,3,5):
    urn["black"]+=1
 elif face==6:
    urn["red"]+=1
 else:
    urn["blue"]+=1
 urn_list=["red"]*urn["red"]+["blue"]*urn["blue"]+["black"]*urn["black"]
 return urn_list[np.random.randint(0,10)]=="red"
```

```python
np.random.seed(12)

true_red=(19/60)
simulation=10000
estimated_red=sum([experiment() for _ in range(simulation)])
estimated_red/=simulation
print("Probabilitatea estimata:",estimated_red)
print("Probabilitatea teoretica:",true_red)
```

In true_red retinem valoarea probabilitatii teoretice ca bila sa fie rosie
Definim Variabila aleatoare X pentru zaruri
R=bila selectata este rosie
P(X=prime)=P(X=2)+P(X=3)+P(X=5)=1/2
P(X=6)=1/6
P(X=others)=P(X=4)+P(X=1)=1/3
Folosind formula probabilitatii totale:
P(R)=P(R|X=6)*P(X=6)+P(R|X=prime)*P(X=prime)+P(R|X=others)*P(X=others)
    =19/60


Exercitiul 2.1:

```python
X1 = stats.poisson.rvs(mu=1, size=1000)
X2 = stats.poisson.rvs(mu=2, size=1000)
X3 = stats.poisson.rvs(mu=5, size=1000)
X4 = stats.poisson.rvs(mu=10, size=1000)
X=[X1,X2,X3,X4]
```

Exercitiul 2.2.a):

```python
lambdas=[1,2,5,10]
l=np.random.choice(lambdas,size=1000,p=[0.25,0.25,0.25,0.25])
X_random=stats.poisson.rvs(mu=l)

for index,x in enumerate(X):
    plt.hist(x,bins=15,color='orange',edgecolor='black',alpha=0.7)
    plt.title(f'Distribuția Poisson (lambda = {lambdas[index]})')
    plt.xlabel('Valori')
    plt.ylabel('Frecvență')
    plt.show()

plt.hist(X_random,bins=15,color='orange',edgecolor='black',alpha=0.7)
plt.title(f'Distribuția Poisson (lambda = random)')
plt.xlabel('Valori')
plt.ylabel('Frecvență')
plt.show()
```

2.2.b) Se observa ca frecventa valorilor mai mici este mai mare in cazul distributiei Poisson randomizate
