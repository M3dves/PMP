```python
import numpy as np
from hmmlearn import hmm
```

a)


```python
states = ["W", "R", "S"]

observations = ["L", "M", "H"]

pi = np.array([0.4, 0.3, 0.3]) #initial probabilities

A = np.array([
    [0.6, 0.3, 0.1],
    [0.2, 0.7, 0.1],
    [0.3, 0.2, 0.5]
]) #transition matrix

B = np.array([
    [0.1, 0.7, 0.2],
    [0.05, 0.25, 0.70],
    [0.8, 0.15, 0.05]
]) #emission matrix

model = hmm.CategoricalHMM(n_components=3, init_params="")
model.startprob_ = pi
model.transmat_ = A
model.emissionprob_ = B


obs_seq = np.array([[1], [2], [0]])
```

b)


```python
log_prob = model.score(obs_seq)
forward_prob = np.exp(log_prob)

print("P([M, H, L]) =", forward_prob)
```

c)Daca am folosi Brute-force ,complexitatea timp ar fi exponentiala(Pentru exemplul nostru cu 3 stari si o secventa de 10000 de observatii ,vom avea nevoie de un numar de operatii de 2 * 10 000 * 3^(10 000),un numar foarte mare).
Algoritmul Viterbi,pentru aceeasi secventa de observatii ,va face 10 000*3^2 oparatii ,


```python
log_prob_vit, state_seq = model.decode(obs_seq, algorithm="viterbi")

decoded_states = [states[s] for s in state_seq]

print("Most likely state sequence:", decoded_states)
print("Log-probability:", log_prob_vit)
```

d)


```python
N = 10000
count = 0

for _ in range(N):
    X, Z = model.sample(3)
    if (X.flatten() == obs_seq.flatten()).all():
        count += 1

empirical_prob = count / N

print("Empirical P([M, H, L]) =", empirical_prob)
print("Exact (Forward) =", forward_prob)
```
